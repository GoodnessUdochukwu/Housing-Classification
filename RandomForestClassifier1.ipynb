##### **Initiate Column Transformer and Build Pipeline with Classifier**

```python
rf_numeric_pipeline = make_pipeline(SimpleImputer())
rf_numeric_pipeline
```

```python
rf_cat_ordinal_pipeline = make_pipeline(SimpleImputer(strategy="constant", fill_value="N_A"),\
                                         OrdinalEncoder(categories=cat_ord_categ))
rf_cat_ordinal_pipeline
```

```python
rf_cat_onehot_pipeline = make_pipeline(SimpleImputer(strategy="constant", fill_value="N_A"),\
                                        OneHotEncoder(drop='first', handle_unknown="ignore"))
rf_cat_onehot_pipeline
```

```python
preprocessor = ColumnTransformer(transformers=[('rf_cat_onehot', rf_cat_onehot_pipeline, X_cat_onehot.columns),
                                               ('rf_num_pipe', rf_numeric_pipeline, X_num.columns),
                                               ('rf_cat_ord', rf_cat_ordinal_pipeline, X_cat_ord.columns),
                                              ]
                                )
preprocessor
```

```python
rf_fullpipeline = make_pipeline(preprocessor, RandomForestClassifier())
rf_fullpipeline
```

```python
param = {'columntransformer__rf_num_pipe__simpleimputer__strategy': ['mean', 'median'],
         'randomforestclassifier__n_estimators': range(2, 10, 2),
         'randomforestclassifier__criterion': ["gini", "entropy"],
         'randomforestclassifier__max_depth': range(2, 15, 2),
         'randomforestclassifier__min_samples_split': range(2, 15, 2),
         'randomforestclassifier__min_samples_leaf': range(2, 20, 2),
         'randomforestclassifier__random_state': [25],
         'randomforestclassifier__class_weight': [None, 'balanced'],
         'randomforestclassifier__bootstrap': [False]
         }
param
```

```python
rf_search = GridSearchCV(rf_fullpipeline, param, cv=5, verbose=1)
rf_search
```

```python
rf_search.fit(X_train, y)
```

```python
rf_search.best_params_
```

```python
rf_search.best_estimator_
```

```python
rf_search.best_score_
```

```python
y_pred = rf_search.predict(X_train)
y_pred
```

```python
accuracy_score(y_pred, y)
```
